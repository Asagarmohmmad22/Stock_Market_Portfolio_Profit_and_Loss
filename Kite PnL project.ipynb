```python
df = df[df.Status != 'CANCELLED']
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.drop([0,2])
```

```python
df['Qty.'] = df['Qty.'].str.split('/').str[0]
```

```python
df
```

```python
df['Qty.'] = df['Qty.'].astype(int)
```

```python
df
```

```python
df['Turnover'] = df['Qty.']*df['Avg. price']
```

```python
df['Turnover'] = df['Turnover'].astype(int)
```

```python
df
```

```python
df['Brokerage'] =  np.where(df['Turnover']*0.03/100 > 20, 20, df['Turnover']*0.03/100)
```

```python
df
```

```python
df['STT/CTT'] = np.where(df['Type']=='SELL', df['Turnover']*0.025/100, 0)
```

```python
df
```

```python
df['ETC'] = df['Turnover']*0.00325/100
```

```python
df
```

```python
df['SEBI'] = df['Turnover'] * (10 / 10000000)+(df['Turnover'] * (10 / 10000000))*18/100
```

```python
df
```

```python
df['GST'] = (df['Brokerage']+df['SEBI']+df['ETC'])*18/100
```

```python
df
```

```python
df['Stamp charges'] = np.where(df['Type']=='BUY', df['Turnover']*0.003/100, 0)
```

```python
df
```

```python
df['Total charges'] = df['Brokerage']+df['STT/CTT']+df['ETC']+df['SEBI']+df['GST']+df['Stamp charges']
```

```python
df
```

# 2) Stock wise and Type wise analysis with weighted Avg. price & calculated charges

```python
stock_type_analy = df.groupby(['Instrument','Type']).agg({'Qty.':'sum', 'Avg. price':'mean','Turnover':'sum','Brokerage':'sum','STT/CTT':'sum','ETC':'sum','SEBI':'sum','GST':'sum','Stamp charges':'sum','Total charges':'sum'})
```

```python
stock_type_analy
```

# 3) Overall Summary of each Stocks


```python
df['Gross PnL'] = np.where((df['Type']=='SELL'),df['Turnover'],-1*df['Turnover'])
df['Gross PnL']
```

```python
df2 = df.groupby(['Instrument'])[['Gross PnL','Total charges']].sum()
df2
```

```python
df2['Net PnL'] = df2['Gross PnL']-df2['Total charges']
df2['Net PnL']
```

```python
df2['% Charges on Gross PnL'] = (df2['Total charges']/df2['Gross PnL'])*100
df2
```

```python
total = df2.sum(axis=0)
total
```

```python
df2.loc['Total'] = df2.sum(axis=0)
```

```python
df2
```

```python
df2.round(2).reset_index()
```

```python
writer = pd.ExcelWriter("Kite_Pnl_Output_File.xlsx")

df.to_excel(writer, sheet_name='individual Transaction Charges',index=False)
stock_type_analy.to_excel(writer, sheet_name='stock_type_analy')
df2.to_excel(writer, sheet_name='Overall Summary of each Stocks')

writer.close() # The save() method has been deprecated and removed in Pandas. You should use close() instead.
```
